```python
#Modify feature list and create new DataFrame
our_features.remove('bedrooms')
our_features += ['host_listings_count', 'has_availability', 'minimum_maximum_nights']
df = df[our_features]
```

```python
#remove missing values
df['beds_na'] = df['beds'].isnull()
mean_beds = round(df['beds'].mean())
df['beds'].fillna(value=mean_beds, inplace=True)
np.sum(df.isnull())
```

```python
#Remove outliers
df['price_win'] = stats.mstats.winsorize(df['price'], limits=[0.01,0.01])
df['review_scores_rating_win'] = stats.mstats.winsorize(df['review_scores_rating'], limits=[0.01,0.01])
#np.unique(df['review_scores_rating'] - df['review_scores_rating_win'])
```

```python
#Check Winsorized categories to examine distribution and determine if outliers still exist
fig1 = sns.histplot(data=df['price_win']) #Skewed relationship
```

```python
#Apply a log transform:
fig2 = sns.histplot(data=np.log1p(df['price_win']))
```

```python
#Convert the 'has_availability' column from bool to int
df['has_availability'] = df['has_availability'].map({True:1, False:0})
np.sum(df.isnull())
```

```python
#One Hot Encode non-numeric features
non_numer_features = df.select_dtypes(include=['object'])
df_room_type = pd.get_dummies(df['room_type'])
df_neighbourhood = pd.get_dummies(df['neighbourhood_group_cleansed'])
df.drop('room_type', axis=1, inplace=True)
df.drop('neighbourhood_group_cleansed', axis=1,inplace=True)
df = df.join(df_room_type)
df = df.join(df_neighbourhood)
```

```python
corrs_sorted = df.corr()['price_win'].sort_values(ascending=False).drop(['price_win', 'price'])
corrs_sorted
```

```python
#Before I train the model, let me drop unnecessary features like 'beds_na' 
#and set the 'price_win' column to 'price'
df.drop('beds_na', axis=1, inplace=True)
df['price'] = df['price_win']
df['review_scores_rating'] = df['review_scores_rating_win']
df.drop('price_win', axis=1, inplace=True)
df.drop('review_scores_rating_win', axis=1, inplace=True)
df.columns
```

```python
#Model Training Time!
#Model 1: SGDRegressor
#Step 1: Select features for X and label for y
X = df.drop('price', axis=1)
y = df['price']

#Step 2: Create training and testing data sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=1234)

scaler = StandardScaler()

scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

#Step 3: Initialize the model and fit to training data
model = LinearRegression()
model.fit(X_train_scaled, np.log1p(y_train))

#Step 3: Make predictions for X_test

model1_pred_log = model.predict(X_test_scaled)
#Since we applied a log transformation, we must compute the inverse before running our evaluation metrics
model1_pred_real = np.expm1(model1_pred_log)
```

```python
#Examine performance of Linear Regressor model with RMSE metric
model1_rmse = np.sqrt(mean_squared_error(y_test, model1_pred_real))
model1_r2 = r2_score(y_test, model1_pred_real)
print("RMSE: {0}".format(model1_rmse))
print("R2: {0}" .format(model1_r2)) #Possible overfitting
```

```python
#Model 2: Gradient Boosted Regressor
model2 = GradientBoostingRegressor(max_depth=10, n_estimators=300)
model2.fit(X_train_scaled, np.log1p(y_train))
```

```python
model2_pred1_log = model2.predict(X_test_scaled)
model2_pred_real =np.expm1(model2_pred1_log)

model2_rmse = np.sqrt(mean_squared_error(y_test, model2_pred_real))
model2_r2 = r2_score(y_test, model2_pred_real)

print("RMSE: {0}".format(model2_rmse))
print("R2: {0}".format(model2_r2))
```

```python
#It looks like we have a winner! Let's run a GridSearchCV to find the best set of hyperparameters for the GBDT.
#Create param_grid of features to modify
n_trees = [100*i for i in range(1,6)]
depths = [i for i in range(3, 10)]

param_grid = {'n_estimators': n_trees,
                'max_depth': depths}

#Create a GridSearchCV object for the model
our_grid = GridSearchCV(model2, param_grid, cv=5)

print("Running grid search...")
grid_result = our_grid.fit(X_train_scaled, np.log1p(y_train))
print("Done")
```

```python
#Optimal hyperparameters for our GBDT model
optimal_max_depth = grid_result.best_params_['max_depth']
optimal_n_estimators = grid_result.best_params_['n_estimators']
print("Best max_depth: " + str(optimal_max_depth))
print("Best n_estimators: " + str(optimal_n_estimators))
```

```python
#Training optimal GBDT model
model2_optimal = GradientBoostingRegressor(max_depth=optimal_max_depth, n_estimators=optimal_n_estimators)
model2_optimal.fit(X_train_scaled, np.log1p(y_train))
```

```python
#Predictions from optimal GBDT model and evaluation

model2_optimal_pred_log = model2_optimal.predict(X_test_scaled)
model2_optimal_pred_real = np.expm1(model2_optimal_pred_log)

model2_optimal_rmse = np.sqrt(mean_squared_error(y_test, model2_optimal_pred_real))
model2_optimal_r2 = np.sqrt(r2_score(y_test, model2_optimal_pred_real))

print("RMSE: {0}".format(model2_optimal_rmse))
print("R2: {0}".format(model2_optimal_r2))
```
